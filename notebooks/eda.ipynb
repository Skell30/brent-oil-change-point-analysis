```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
try:
    brent=pd.read_csv(r'C:\Users\pc\brent-oil-change-point-analysis\data\brent.csv')
    print('load successful')
except:
    print('try again')    
```

```python
brent.head()
```

```python
# Plot the Brent price over time
plt.figure(figsize=(15, 5))
plt.plot(brent['Date'], brent['Price'], color='navy')
plt.title('Brent Oil Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD/barrel)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate daily log returns
brent['LogReturn'] = np.log(brent['Price'] / brent['Price'].shift(1))

# Drop NaN created by shift
brent.dropna(inplace=True)
```

```python
# Plot log returns
plt.figure(figsize=(15, 4))
plt.plot(brent['Date'], brent['LogReturn'], color='darkred')
plt.title('Log Returns of Brent Oil Prices')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram of log returns
plt.figure(figsize=(7, 4))
sns.histplot(brent['LogReturn'], bins=100, kde=True)
plt.title('Distribution of Log Returns')
plt.xlabel('Log Return')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Volatility clustering view
plt.figure(figsize=(15, 4))
plt.plot(brent['Date'], brent['LogReturn'].rolling(30).std(), label='30-day Rolling Std Dev')
plt.title('Volatility Clustering in Log Returns')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
